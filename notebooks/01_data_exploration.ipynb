```python
import pandas as pd
import numpy as np

# 1. Load CSV
df = pd.read_csv('../data/raw/student_performance.csv')

# 2. Display first 5 rows
print("First 5 rows:")
print(df.head())
print("\n" + "="*50 + "\n")

# 3. Show column names
print("Column names:")
print(df.columns.tolist())
print("\n" + "="*50 + "\n")

# 4. Show data types
print("Data types:")
print(df.dtypes)
print("\n" + "="*50 + "\n")

# 5. Check missing values
print("Missing values per column:")
print(df.isnull().sum())
print("\nTotal missing values:", df.isnull().sum().sum())
print("\n" + "="*50 + "\n")

# 6. Show basic shape info
print("Shape info:")
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
print(f"Total entries: {df.size}")
print("\n" + "="*50 + "\n")

# Bonus: Show some basic statistics
print("Basic statistics for numeric columns:")
print(df.describe())
```
